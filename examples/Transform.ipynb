# Check Array Transposer

```python
d = np.random.random(1000).reshape((50, 10, 2))
t = wtools.transpose(d)
np.allclose(np.flip(d.swapaxes(0, 1), 2), t)
```

```python
t = wtools.transpose(t)
np.allclose(d, t)
```

# Display functions

First lets display the the `d` ndarray we made that has a shape of `(50, 10, 2)`. Note that `wtools.display` ensures that the first dimension (`50`) passed lies on the X-axis and the second dimension (`10`) passed lies on the Y-axis int positive up direction.

```python
wtools.display(plt, d[:, :, 0])
plt.title('What we expect in a cartesian space')
plt.show()
```

This is quite different from `matplotlib`'s default behaviour for plotting images:

```python
# NOT WHAT WE EXPECT
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.pcolormesh(d[:, :, 0])
plt.title('plt.pcolormesh')

plt.subplot(1,2,2)
plt.imshow(d[:, :, 0])
plt.title('plt.imshow')

plt.tight_layout()
plt.show()
```
